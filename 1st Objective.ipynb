```python
df['Property Type'].value_counts()
```

```python
df['Property Type'].value_counts()
```

```python
july_sales.groupby(['Sale Date','Property Type'])['Property Type'].count()
#not proper structure data frame
# we have to make it more structure so we use unstack
```

```python
daily_sales_july_prop=july_sales.groupby(['Sale Date','Property Type'])['Property Type'].count().unstack()
daily_sales_july_prop
```

```python
plt.rcParams['figure.figsize']=(18,5) #rcParams is used to make the legend 
daily_sales_july_prop.plot()
plt.xlabel('Day')
plt.ylabel('Total number of Sales')
plt.title('Daily Sales Volume in July 2018')
```

```python
#box plot
```

```python
july_prop_plot=july_sales.groupby(['Sale Date','Property Type'])['Price'].mean().unstack()
july_prop_plot
```
