```python
df['ShortPassing']=df['ShortPassing'].fillna(df['ShortPassing'].mode()[0])
#[0] is used to access the first (most frequent) value from the mode calculation
```

```python
df.dtypes
```

```python
for col in df.columns:
   df[col].fillna(df[col].mode()[0],inplace=True)
  
#use if else for large data types otherwise it can be written directly too
#if there are any null values in object it changes it with mode()
    
```

```python
df.isnull().sum().sum()
#sum of all null values not columns
```

```python
import matplotlib.pyplot as plt
labels=df['Preferred Foot'].value_counts().index
#index value is stored in foot
labels
```

```python
sizes=df['Preferred Foot'].value_counts().values
sizes
```

```python
myexplode=[0.1,0]
mycolor=['blue','green']
plt.pie(sizes,labels=labels,autopct='%0.3f%%',startangle=90,explode = myexplode,colors=mycolor)
plt.legend()
plt.title('Distribution of Preferred Foot')
plt.show()
```

```python
#What is the international reputation of left and right footed players? print it with their count
```

```python
foot=df.groupby('Preferred Foot')['International Reputation'].value_counts().unstack()
foot
#way1 of handling
```

```python
foot=df.groupby(['Preferred Foot','International Reputation'])['International Reputation'].count().unstack()
foot
#easier/another way of handling
```

```python
plt.figure(figsize=(20,5))
foot.plot(kind='bar')
plt.xlabel('Preferred Foot')
plt.ylabel('Count')
plt.title('Internation Reputation of Left and Right Footed Players')
plt.show()
```
